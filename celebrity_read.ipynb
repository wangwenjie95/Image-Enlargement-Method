```python
import os
import numpy as np
from PIL import Image


path = 'C:/Users/admin/Desktop/img_align_celeba/'

def readpicture():
    inputt8 = []
    outputt32 = []
    
    n = 1
    
    image_paths = [os.path.join(path, f) for f in os.listdir(path)]
    for image_path in image_paths:
        if(image_path!='C:/Users/admin/Desktop/img_align_celeba/Thumbs.db'):
            print n, 
            image_pil = Image.open(image_path)
            outt32 = image_pil.resize((32,32))
            inn8 = outt32.resize((8,8))
        
        
            iii = np.array(inn8, 'uint8')
            inputt8.append(iii)
        
            ooo = np.array(outt32, 'uint8')
            outputt32.append(ooo)
            n = n+1
    
    inputt8 = np.array(inputt8)
    outputt32 = np.array(outputt32)
    np.save('input8_8.npy',inputt8)
    np.save('output32_32.npy',outputt32)

readpicture()
```
